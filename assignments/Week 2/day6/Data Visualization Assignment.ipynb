# Data Visualization assignment

This one is simple, you have to plot a map of population by country for the year 2016.

For this you have to find:

- Population data 
- A country level geojson file.

You can use either one of the map plotting libraries we used in class.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import json
import folium
```

```python
pop_data = pd.read_csv("data/bitch.csv")
pop_data2016 = pop_data[['Country Name','Country Code','2016']]
pop_data2016
```

```python
pop_data2016.dtypes
```

```python
geo_data["features"][0]["properties"]['name']
```

```python
with open("data/custom.geo.json") as fname:
    geo_data = json.load(fname)
```

```python
m = folium.Map()
m.choropleth(
    geo_data=geo_data,
    name='world pop data',
    data=pop_data2016,
    columns=["Country Name", "2016"],
#    threshold_scale=[0, 1000000, 10000000, 100000000, 1000000000, 2000000000],
    key_on="feature.properties.name",
    fill_color="YlOrRd",
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name="Population by country in 2016",
)

folium.LayerControl().add_to(m)
m
```
